**Portfolio optimization using FinQuant Library**

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Ticker symbols for VOO and QQQM
voo_ticker = 'VOO'
qqqm_ticker = 'QQQM'

# Download data for VOO and QQQM from Yahoo Finance
voo_data = yf.download(voo_ticker, start='2023-01-01', end='2023-11-15')
qqqm_data = yf.download(qqqm_ticker, start='2023-01-01', end='2023-11-15')

# Calculate normalized prices based on the first available price
voo_normalized = voo_data['Adj Close'] / voo_data['Adj Close'].iloc[0] * 100
qqqm_normalized = qqqm_data['Adj Close'] / qqqm_data['Adj Close'].iloc[0] * 100

# Plotting VOO and QQQM normalized prices on the same plot
plt.figure(figsize=(10, 5))
plt.plot(voo_normalized, label='VOO')
plt.plot(qqqm_normalized, label='QQQM', color='orange')
plt.title('Rebased NYSEARCA: VOO vs NASDAQ: QQQM Stock Prices')
plt.xlabel('Date')
plt.ylabel('Normalized Price (Base 100)')
plt.legend()
plt.show()
```

```python
import datetime
import pandas as pd
from finquant.portfolio import build_portfolio

d = {

    1: {"Name": "NLY", "Allocation": 20},
    2: {"Name": "LVS", "Allocation": 20},
    3: {"Name": "FXAIX", "Allocation": 20},
    4: {"Name": "PG", "Allocation": 20},
    5: {"Name": "CAT", "Allocation": 15},
    6: {"Name": "HPP", "Allocation": 5},
}

pf_allocation = pd.DataFrame.from_dict(d, orient="index")

names = pf_allocation["Name"].values.tolist()
start_date = datetime.datetime(2020, 1, 1)
end_date = "2023-11-01"
market_index = "^GSPC"

pf = build_portfolio(
    names=names,
    pf_allocation=pf_allocation,
    start_date=start_date,
    end_date=end_date,
    data_api="yfinance",
    market_index=market_index,
)

print(pf.portfolio)
print(pf.data.head(3))
print(pf)
pf.properties()
```

```python
pf.properties()
```

```python
import datetime
import matplotlib.pyplot as plt
from finquant.moving_average import sma, ema, compute_ma, plot_bollinger_band

# Annualized mean returns
print(pf.comp_mean_returns())

# Daily returns (percentage change)
print(pf.comp_cumulative_returns().head(3))
print(pf.comp_daily_log_returns().head(3))

# Plotting stock data of the portfolio
pf.data.plot()
plt.show()

# Plotting with a secondary y-axis for better visualization
pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True)
plt.show()

# Plotting cumulative returns
pf.comp_cumulative_returns().plot().axhline(y=0, color="black", lw=3)
plt.show()

# Plotting daily percentage changes of returns
pf.comp_daily_returns().plot().axhline(y=0, color="black")
plt.show()

# Plotting daily log returns
pf.comp_daily_log_returns().plot().axhline(y=0, color="black")
plt.show()

# Plotting cumulative log returns
pf.comp_daily_log_returns().cumsum().plot().axhline(y=0, color="black")
plt.show()

# Moving Averages

# Simple Moving Average
ax = pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True)
sma(pf.data, span=50).plot(ax=ax, secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True)
plt.show()

# Exponential Moving Average
ax = pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True)
ema(pf.data).plot(ax=ax, secondary_y=["WIKI/NLY", "WIKI/LVS"])
plt.show()

# Band of moving averages and Buy/Sell signals
spans = [10, 50, 100, 150, 200]

# Replace 'WIKI/LVS' with the correct Yahoo Finance symbol for Las Vegas Sands Corp.
lvs_symbol = "LVS"

# Retrieve the correct stock data for LVS
lvs = pf.get_stock(lvs_symbol).data.copy(deep=True)
compute_ma(lvs, ema, spans, plot=True)
plt.show()

# Plot the Bollinger Band of one stock
span = 20

# For simple moving average:
plot_bollinger_band(lvs, sma, span)
plt.show()

# For exponential moving average:
plot_bollinger_band(lvs, ema, span)
plt.show()

# Recomputing expected return, volatility, and Sharpe ratio
pf.freq = 100
pf.risk_free_rate = 0.02
exret = pf.comp_expected_return(freq=100)
vol = pf.comp_volatility(freq=100)
sharpe = pf.comp_sharpe()
print(f"For {pf.freq} trading days and a risk-free rate of {pf.risk_free_rate}:")
print(f"Expected return: {exret:.3f}")
print(f"Volatility: {vol:.3f}")
print(f"Sharpe Ratio: {sharpe:.3f}")

# Portfolio properties
pf.properties()

# Extracting data of stocks individually
lvs = pf.get_stock("LVS")
print(lvs.data.head(3))
print(lvs.comp_daily_returns().head(3))
print(lvs.expected_return)
print(lvs.volatility)
print(lvs.skew)
print(lvs.kurtosis)
print(lvs)
lvs.properties()

# Extracting stock data by date
print(pf.data.loc[str(datetime.datetime(2020, 1, 2))])
print(pf.data.loc[pf.data.index > datetime.datetime(2021, 1, 2)].head(3))
print(pf.data.loc[pf.data.index.year == 2022].head(3))
```

```python
import datetime
import matplotlib.pyplot as plt
from finquant.moving_average import sma, ema, compute_ma, plot_bollinger_band

# Create a 4x3 subplot grid
fig, axs = plt.subplots(4, 3, figsize=(20, 15))

# Annualized mean returns
axs[0, 0].set_title('Annualized Mean Returns')
pf.comp_mean_returns().plot(kind='bar', ax=axs[0, 0])

# Daily returns (percentage change)
axs[0, 1].set_title('Daily Returns (Percentage Change)')
pf.comp_cumulative_returns().head(3).plot(kind='bar', ax=axs[0, 1])

# Plotting stock data of the portfolio
axs[0, 2].set_title('Stock Data of the Portfolio')
pf.data.plot(ax=axs[0, 2])

# Plotting with a secondary y-axis for better visualization
axs[1, 0].set_title('Stock Data with Secondary Y-axis')
pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True, ax=axs[1, 0])

# Plotting cumulative returns
axs[1, 1].set_title('Cumulative Returns')
pf.comp_cumulative_returns().plot(ax=axs[1, 1])
axs[1, 1].axhline(y=0, color="black", lw=3)

# Plotting daily percentage changes of returns
axs[1, 2].set_title('Daily Percentage Changes of Returns')
pf.comp_daily_returns().plot(ax=axs[1, 2])
axs[1, 2].axhline(y=0, color="black")

# Plotting daily log returns
axs[2, 0].set_title('Daily Log Returns')
pf.comp_daily_log_returns().plot(ax=axs[2, 0])
axs[2, 0].axhline(y=0, color="black")

# Plotting cumulative log returns
axs[2, 1].set_title('Cumulative Log Returns')
pf.comp_daily_log_returns().cumsum().plot(ax=axs[2, 1])
axs[2, 1].axhline(y=0, color="black")

# Simple Moving Average
axs[2, 2].set_title('Simple Moving Average')
ax = pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True, ax=axs[2, 2])
sma(pf.data, span=50).plot(ax=ax, secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True)

# Exponential Moving Average
axs[3, 0].set_title('Exponential Moving Average')
ax = pf.data.plot(secondary_y=["WIKI/NLY", "WIKI/LVS"], grid=True, ax=axs[3, 0])
ema(pf.data).plot(ax=ax, secondary_y=["WIKI/NLY", "WIKI/LVS"])

# Band of moving averages and Buy/Sell signals
axs[3, 1].set_title('Moving Averages and Buy/Sell Signals')
spans = [10, 50, 100, 150, 200]
lvs_symbol = "LVS"
lvs = pf.get_stock(lvs_symbol).data.copy(deep=True)
lvs_ma = compute_ma(lvs, ema, spans)
lvs_ma.plot(ax=axs[3, 1])
axs[3, 1].legend(spans)  # Adding legend

# Plot the Bollinger Band of one stock
axs[3, 2].set_title('Bollinger Band')
span = 20
plot_bollinger_band(lvs, sma, span)  # No ax argument here

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import numpy as np
from finquant.efficient_frontier import EfficientFrontier
# Portfolio optimisation
print("pf.risk_free_rate = {}".format(pf.risk_free_rate))
print("pf.freq = {}".format(pf.freq))

# Efficient Frontier
pf.ef_minimum_volatility(verbose=True)
pf.ef_maximum_sharpe_ratio(verbose=True)
pf.ef_efficient_return(0.26, verbose=True)
pf.ef_efficient_volatility(0.22, verbose=True)

# Manually creating an instance of EfficientFrontier
ef = EfficientFrontier(pf.comp_mean_returns(freq=1), pf.comp_cov())
print(ef.minimum_volatility())
(expected_return, volatility, sharpe) = ef.properties(verbose=True)

# Computing and visualising the Efficient Frontier
pf.ef_plot_efrontier()
pf.ef_plot_optimal_portfolios()
pf.plot_stocks()
plt.show()

# Manually creating an Efficient Frontier with target return values
targets = np.linspace(0.12, 0.45, 50)
efficient_frontier = ef.efficient_frontier(targets)
ef.plot_efrontier()
pf.ef.plot_optimal_portfolios()
pf.plot_stocks()
plt.show()

# Monte Carlo
opt_w, opt_res = pf.mc_optimisation(num_trials=5000)
pf.mc_properties()
pf.mc_plot_results()
pf.plot_stocks()
plt.show()

# Optimisation overlay
opt_w, opt_res = pf.mc_optimisation(num_trials=5000)
pf.mc_plot_results()
pf.ef_plot_efrontier()
pf.ef.plot_optimal_portfolios()
pf.plot_stocks()
plt.show()
```
